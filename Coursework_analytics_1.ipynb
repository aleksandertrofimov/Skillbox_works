```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib
from matplotlib import pyplot as plt
import math

# Чтение courses.csv. Переименование столбца 'id' в 'course_id' и удаление столбца 'Unnamed: 0'
df = pd.read_csv('data/courses.csv').rename(columns={'id' : 'course_id', 'title' : 'course_title'}).drop(
    'Unnamed: 0', axis=1)

# Чтение course_contents.csv
df_additional_1 = pd.read_csv('data/course_contents.csv') 

#Чтение progresses.csv. Переименование столбца 'id' в 'progress_id'
df_additional_2 = pd.read_csv('data/progresses.csv').rename(columns={'id' : 'progress_id'}) 

#Чтение progress_phases.csv.
df_additional_3 = pd.read_csv('data/progress_phases.csv')

#Наибольшее количество строк в progress_phases.csv, далее по убыванию progresses.csv, course_contents.csv и courses.csv 
#В силу этого обстоятельства и пересечений колонок объединять буду сначала попарно courses.csv и course_contents.csv
#по 'course_id' и progresses.csv и progress_phases.csv по 'progress_id', а затем получившиеся результаты по 
#колонкам 'course_id', 'module_number', 'lesson_number'.

# Объединение courses.csv и course_contents.csv
df = df.merge(df_additional_1, how='outer', on='course_id')

# Объединение progress_phases.csv и progresses.csv
df_additional_3 = df_additional_3.merge(df_additional_2, how='outer', on='progress_id')

# Объединение в конечный датасет
df = df.merge(df_additional_3, how='outer', on=['course_id', 'module_number', 'lesson_number'])

# Исправление типов данных в столбцах номеров модулей и файлов (NAN при этом заменяются на 0)

df['module_number'] = df['module_number'].fillna(0).astype('int8')
df['lesson_number'] = df['lesson_number'].fillna(0).astype('int8')
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce', yearfirst=True, utc=True)
df['finish_date'] = pd.to_datetime(df['finish_date'], errors='coerce', yearfirst=True, utc=True)
df.info()
# проверка основного датасета
df[df.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

```python
#общее количество курсов в датасете,
course_quantity = len(df['course_id'].unique())
print('Общее количество курсов ' + str(course_quantity))
```

```python
#количество модулей на каждом курсе

#Расчёт количества модулей для каждого курса.
#Cрез по колонкам 'course_id' и 'module_title', затем удаление дубликатов, 
# группировка по столбцу 'course_id' и подсчет значений по 'module_title'
module_count = df.loc[:, ['course_id', 'module_title']].drop_duplicates().groupby('course_id')['module_title'].count()
#Оформрление результата
#датасет с названиями курса
course_search = df.loc[:, ['course_id','course_title']].drop_duplicates('course_id') 
#объединение датасета с названиями курса и расчёта количества модулей
course_search = course_search.merge(module_count, how='outer',on='course_id') 
#актуализация названий столбцов
course_search.columns = ['course_id','course_title', 'module_quantity']
course_search
```

```python
#количество уроков в каждом модуле на каждом курсе,

#Расчёт количества уроков для каждого модуля каждого курса.
#Cрез по колонкам 'course_id', 'course_title', 'module_number','module_title', 'lesson_token', 
#затем удаление дубликатов и NAN, 
# группировка по столбцам 'course_id' и 'module_number' и подсчет значений по 'module_title'

#Формирование датасета
lesson_search = df.loc[:,[
    'course_id', 'course_title', 'module_number','module_title', 'lesson_token']].dropna().drop_duplicates()
#Расчет количества уроков
lesson_count = lesson_search.groupby(['course_id', 'module_number'])['lesson_token'].count()

#Оформление результатов
#Объединение расчёта с исходным датасетом
lesson_search = lesson_search.merge(lesson_count, how='inner',on=['course_id', 'module_number']).drop(
    'lesson_token_x',axis=1).rename(columns={'lesson_token_y': 'lesson_quantity'})
#Удаление дубликатов
lesson_search.drop_duplicates(inplace=True, ignore_index=True)
lesson_search
```

```python
#медианное количество уроков в модуле на каждом курсе,
#Расчет
lesson_median = lesson_search.groupby('course_id')['lesson_quantity'].median()
#Оформление (объединение расчёта с датасетом)
course_search = course_search.merge(lesson_median, how='outer',on='course_id').rename(
    columns={'lesson_quantity':'lessons_in_module_median'})
course_search
```

```python
#количество учеников на каждом курсе

#Расчет количества учеников и переименование столбца 'student_id' в 'student_quantity'
students_search = df.dropna(subset=['student_id']).groupby('course_id')['student_id'].nunique()
#Оформление результатов (срез базового датасета по колонкам 'course_id'и 'course_title', 
#очистка от дубликатов и NAN, объединение с результатами, актуализация названий колонок)
students_search = df.loc[:,['course_id','course_title']].dropna().drop_duplicates().merge(
    students_search, how='outer', on='course_id').rename(columns={'student_id':'student_quantity'})
students_search.sort_values(by=['course_title'], inplace=True, ignore_index=True)
students_search
```

```python
#минимальный, максимальный, средний, медианный возраст студентов
#Чтение данных по студентам (students.csv)
df_students = pd.read_csv('data/students.csv').rename(columns={'id' : 'student_id'}).drop('id_', axis=1)
df_students['birthday'] = pd.to_datetime(
    df_students['birthday'], errors='coerce', yearfirst=True, utc=True)
#Присоединение данных по началу обучения
df_date = df.groupby('student_id')['start_date'].min()
df_students = df_students.merge(df_date, how='outer',on='student_id')
#Вычисление возраста на начало обучения
df_students['age'] = df_students['start_date'] - df_students['birthday']
df_students['age'] = df_students['age'].astype('timedelta64[Y]')
#Вычисление минимального, максимального, среднего, медианного возрастов студентов
df_students_age_min = int(df_students['age'][14 < df_students['age']].min())
df_students_age_max = int(df_students['age'][df_students['age'] < 80].max())
df_students_age_mean = int(df_students['age'][(14 < df_students['age']) & (df_students['age'] < 80)].mean())
df_students_age_median = int(df_students['age'][(14 < df_students['age']) & (df_students['age'] < 80)].median())
#Вывод результатов
print('Минимальный возраст среди студентов ', str(df_students_age_min))
print('Максимальный возраст среди студентов ', str(df_students_age_max))
print('Средний возраст среди студентов ', str(df_students_age_mean))
print('Медианный возраст среди студентов ', str(df_students_age_median))
df_students
```

```python
#минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
#Чтение данных по студентам (students.csv)
df_students = pd.read_csv('data/students.csv').rename(columns={'id' : 'student_id'}).drop('id_', axis=1)
df_students['birthday'] = pd.to_datetime(
    df_students['birthday'], errors='coerce', yearfirst=True, utc=True)
#Присоединение данных по началу обучения
df_date = df.loc[:,['student_id','course_id','start_date']].dropna().drop_duplicates()
df_students = df_students.merge(df_date.loc[:,['student_id','course_id']], how='outer',on='student_id')
df_date = df_date.groupby(['student_id','course_id'])['start_date'].min()
df_students = df_students.merge(df_date, how='outer',on='student_id')
#Вычисление возраста на начало обучения
df_students['age'] = df_students['start_date'] - df_students['birthday']
df_students['age'] = df_students['age'].astype('timedelta64[Y]')
#Вычисление минимального, максимального, среднего, медианного возрастов студентов по каждому курсу
df_students = df_students[(14 < df_students['age']) & (df_students['age'] < 80)].loc[:, ['course_id', 'age']]
df_students_age_min = df_students.groupby('course_id')['age'].min().rename({'age' : 'minimal_age'})
df_students_age_max = df_students.groupby('course_id')['age'].max().rename({'age' : 'maximal_age'})
df_students_age_mean = df_students.groupby('course_id')['age'].mean().rename({'age' : 'mean_age'})
df_students_age_median = df_students.groupby('course_id')['age'].median().rename({'age' : 'median_age'})
#Вывод результатов
df_students = df.loc[:,['course_id', 'course_title']].dropna().drop_duplicates()
for (age, name) in [
    (df_students_age_min, 'minimal_age'),
    (df_students_age_max, 'maximal_age'),
    (df_students_age_mean, 'mean_age'),
    (df_students_age_median, 'median_age')]:
    df_students = df_students.merge(age, how='outer',on='course_id').rename(columns={'age' : name})
    df_students[name] = df_students[name].astype('int32')

df_students.sort_values(by=['course_title'], inplace=True, ignore_index=True)
df_students
```

```python
#2. **(m)** Постройте bar-chart, отражающий количество студентов на каждом курсе. 
#Ticks нужно развернуть так, чтобы они были читаемы

plt.figure(figsize=(10, 5))
#Построение графика
plt.bar(students_search.course_title, students_search.student_quantity) #Построение графика
plt.xlabel('Название курса') #Наименование оси X
plt.ylabel('Количество студентов') #Наименование оси Y
plt.title('Распределение студентов по курсам') #Наименование титула графика
plt.xticks(students_search.course_title, labels=students_search.course_title, rotation='vertical') #Задание меток по оси X
plt.yticks(range(0, 2500, 200), labels=range(0, 2500, 200)) #Задание меток по оси Y
plt.show()
```

```python
#3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, 
#отражающий количество студентов на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. 
#Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). 
#Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. 
#Рамки у графика быть не должно ⭐

#Добавление информации о сфере курса
field_data = df.loc[:,['course_id', 'field']].drop_duplicates().dropna()
students_search = students_search.merge(field_data, how='left', on='course_id')
median_for_course = students_search['student_quantity'].median()
median_label = str('Медиана({:.0f} чел.)'.format(median_for_course))
#Построение графика
plt.figure(figsize=(10, 5))
plt.axes(frame_on=False)
for name, group in students_search.groupby('field'):
    plt.barh(group['course_title'], group['student_quantity'], label=name, alpha=0.1)

plt.axvline(x=median_for_course, ls='--', color='g', label=median_label) #Построение линии медианы
plt.xlabel('Количество студентов') #Наименование оси X
plt.ylabel('Название курса') #Наименование оси Y
plt.title('Распределение студентов по курсам') #Заголовок графика
plt.yticks(students_search.course_title, labels=students_search.course_title) #название столбцов
plt.xticks(range(0, 2500, 200), labels=range(0, 2500, 200)) #метки по оси X
 
plt.legend() #Вывод легенды
plt.show()
```

```python
# 1. **(p)** Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от 
# марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней 
# работы.

#Срез основного сета по 'course_id', 'student_id', 'start_date' при выполнии условия, что в столбце 'is_homework' True, 
#удаление NAN, группировка по курсам и студентам и нахождение минимальной даты в 'start_date' (момент начала обучения) 
students_increase = df[df.is_homework == True].loc[:,['course_id', 'student_id', 'start_date']].dropna().groupby([
    'course_id', 'student_id'])['start_date']. min().reset_index()
# Введение столбца с месяцем начала ('month')
students_increase['month'] = students_increase['start_date'].dt.to_period('M')

# Подсчет дат начала обучения в каждом месяце. Актуализация названия столбца с количеством новых студентов
students_increase = pd.DataFrame(students_increase.groupby([
    'course_id', 'month'])['start_date'].count()).rename(columns={'start_date' : 'student_increase'})
students_increase.reset_index(inplace=True)
# Создание датафрейма с полным рядом всех месяцев за требуемый период для заполнения месяцев без прироста студентов
month_dt = pd.date_range('2016-03-01', '2019-07-31', freq='M')
month_range = pd.DataFrame(month_dt).rename(columns={0 : 'month'})
month_range['month'] = month_range['month'].dt.to_period('M')
# Создание датафрейма со столбцама 'course_id' и 'course_title', удаление повторов и пропущеных значений. 
course_range = df.loc[:,['course_id', 'course_title']].drop_duplicates().dropna().sort_values(
    by=['course_title'], ignore_index=True)
# Объединение всех комбинаций месяцев и курсов.
df2 = pd.DataFrame()
for i in range(len(course_range)):
    df1 = month_range
    df1['course_id'] = course_range.loc[i,'course_id']
    df1['course_title'] = course_range.loc[i,'course_title']
    df2 = df2.append(df1, ignore_index=True)
month_range = df2
# Объединение всех месяцев и курсов с приростом студентов. В случае отсутствия данных о приросте ставится 0.
students_increase = students_increase.merge(month_range, how='outer', on=['course_id', 'month']).fillna(
    {'student_increase' : 0})
# Изменение типа на 'int32' и сортировка
students_increase['student_increase'] = students_increase['student_increase'].astype('int32')
students_increase.sort_values(by=['course_title', 'month'], inplace=True, ignore_index=True)

students_increase
```

```python
# 2. **(m)** На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса.
# 15 графиков. Графики должны иметь заголовки, оси должны быть подписаны. Ticks нужно развернуть так, чтобы они были 
# читаемы.

# Список курсов
course_for_plot = students_increase.loc[:,['course_id', 'course_title']].drop_duplicates().sort_values(
    by=['course_title'], ignore_index=True)
# прореженный массив дат для отметок на графиках
dates = students_increase[(students_increase.index <= 40) & ((students_increase.index - 1) % 3 == 0)].drop_duplicates(
    ignore_index=True).month

# Графики оформляю в 3 столбца и 5 строк
fig, axes = plt.subplots(int(course_for_plot['course_id'].count()/3), 3, figsize=(27,45))
# Генерация графиков по строкам и столбцам
for row in range(int(course_for_plot['course_id'].count()/3)):
    for column in range(3):
        index_course = row*3 + column
        # Формирование датасета для конкретного курса
        increase_data = students_increase[students_increase.course_id == course_for_plot.course_id[index_course]]
        #Задание данных для графика
        axes[row, column].plot(increase_data.month.dt.strftime("%b %Y"), increase_data.student_increase)
        #Титул графика
        axes[row, column].set_title('Прирост студентов на курсе "{}"'.format(course_for_plot.course_title[index_course]))
        #Название осей графика
        axes[row, column].set_xlabel('Даты')
        axes[row, column].set_ylabel('Прирост студентов, чел.')
        #Задание сетки по оси X и надписей отметок
        axes[row, column].set_xticks(dates.dt.strftime("%b %Y"))
        axes[row, column].set_xticklabels(dates.dt.strftime("%b %Y"), rotation_mode='default', rotation=50)
        #Задание сетки по оси  Y и надписей отметок
        yticks = range(0, math.ceil(
            max(increase_data.student_increase)/10+1)*10, math.ceil(max(increase_data.student_increase)/100)*10)
        axes[row, column].set_yticks(yticks)
        axes[row, column].set_yticklabels(yticks)
        #Отрисовка сетки
        axes[row, column].grid()
plt.show()
```

```python
# 3. **(m)** На основании первого пункта построить line-graph с несколькими линиями, отражающими прирост студентов в 
# каждом месяце для каждого курса. 15 линий на графике. Ticks нужно развернуть так, чтобы они были читаемы. График 
# должен иметь заголовок. Ось, отражающая прирост, должна быть подписана. Линия для каждого курса должна иметь свой 
# цвет (нужна легенда). Рамок у графика быть не должно ⭐
fig = plt.figure(figsize=(18,10))
plt.axes(frame_on=False)
# Генерация графиков
for index_course in range(students_increase['course_id'].unique().size):
    increase_data = students_increase[students_increase['course_id'] == course_for_plot.course_id[index_course]]
    #Задание данных для графика
    plt.plot(increase_data.month.dt.strftime("%b %Y"), increase_data.student_increase, label=course_for_plot.course_title[
        index_course])
#Титул графика
plt.title('Прирост студентов по месяцам в зависимости от курса')
#Название осей графика
plt.xlabel('Даты')
plt.ylabel('Прирост студентов, чел.')
#Задание сетки по оси X и надписей отметок
plt.xticks(dates.dt.strftime("%b %Y"))
#Задание сетки по оси  Y и надписей отметок
yticks = range(0, math.ceil(
    max(students_increase.student_increase) / 10) * 10 + 10, 20)
plt.yticks(yticks)
#plt.yticklabels(yticks)
#Отрисовка сетки
plt.grid()
plt.legend(loc='best', title='Название курса', ncol = 1)  #Вывод легенды            
plt.show()
```

```python
# 4. **(p)** Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю (каждый 
# месяц в диапазоне от марта 2016 до июля 2019 включительно) для каждого курса. Учитывать, что выполнение домашнего 
# задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, 
# которые покрывает срок выполнения этих дз)

# создание рабочего датасета путем среза основного сета по 'course_id', 'course_title', 'progress_id', 'start_date' и
#'finish_date' при условии наличия домашнего задания
progresses_slice = df[df['is_homework'] == True].loc[
    :,['course_id', 'course_title', 'progress_id', 'start_date', 'finish_date']]
# Удаление NAN в колонках 'course_title', 'progress_id' и 'start_date'
progresses_slice.dropna(subset=['course_title', 'progress_id', 'start_date'], inplace=True)
# Заполнение NAT в 'finish_date' 1-м августа 2019 года (первый день за пределом требуемого диапазона)
progresses_slice.fillna(pd.to_datetime('2019-08-01 00:00:00+00:00'), inplace=True)
# Сброс индексов после удалений.
progresses_slice.reset_index(inplace=True)
# Конвертация дат в периоды
progresses_slice['start_date'] = progresses_slice['start_date'].dt.to_period('M')
progresses_slice['finish_date'] = progresses_slice['finish_date'].dt.to_period('M')

#Для учета заданий, которые делались в нескольких месяцах объединяю рабочий сет с сетом со всеми месяцами, оставляю 
#данные удовлетворяющие условию "Дата начала"<="месяц"<="Дата окончания". После этого группирую их по курсу и месяцу и 
#считаю количество прогрессов. Чтобы восполнить линейку месяцов снова объединяю с датасетом с набором месяцов и курсов

# Слияние рабочего датасета с датасетом с полным набором курсов и месяцов
progresses_slice = progresses_slice.merge(month_range, how='outer', on=['course_id', 'course_title'])
# Сохранение данных, где месяц находится между датой начала и окончания 
progresses_slice = progresses_slice[progresses_slice['start_date'] <= progresses_slice['month']]
progresses_slice = progresses_slice[progresses_slice['finish_date'] >= progresses_slice['month']]
# Подсчёт прогрессов в каждом месяце
progresses_slice = progresses_slice.groupby(['course_id','course_title', 'month']).index.count()
# Добавление линейки курсов и месяцев 
progresses_slice = month_range.merge(progresses_slice, how='left', on=['course_id', 'course_title', 'month'])
# Заполнение отсутствующих данных по прогрессам 0. Изменение типа на целочисленный
progresses_slice['index'] = progresses_slice['index'].fillna(0).astype('int32')
# Актуализация названия столбца с количеством домашних заданий. 
progresses_slice = progresses_slice.rename(columns={'index' : 'homework_count'}).sort_values(
    by=['course_title', 'month'], ignore_index=True)
progresses_slice
```

```python
# 1. **(p)** Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля (разность между 
# временем начала и окончания выполнения домашней работы) для каждого курса. Если домашних заданий в модуле несколько, 
# то считать разность между временем начала выполнения первой домашней работы и временем окончания выполнения последней 
# домашней работы в модуле

# создание рабочего датасета путем среза основного сета по 'course_id', 'course_title', 'module_number', 'progress_id', 
#'start_date' и 'finish_date' при условии наличия домашнего задания
module_df = df[df['is_homework'] == True].loc[
    :,['course_id', 'course_title', 'module_number', 'start_date', 'finish_date', 'progress_id']]
# Удаление NA 
module_df.dropna(subset=['course_title', 'module_number', 'start_date', 'finish_date'], inplace=True)
# Расчет начала первого домашнего задания в модуле (начала модуля)
module_start = module_df.groupby(['progress_id', 'course_title', 'module_number'])['start_date'].min()
# Расчет окончания последнего домашнего задания в модуле (конца модуля)
module_finish = module_df.groupby(['progress_id', 'course_title', 'module_number'])['finish_date'].max()
# Объединение начала и конца модулей в единый датасет
module_df = pd.DataFrame(module_start).merge(module_finish, how='outer', on=[
    'progress_id', 'course_title', 'module_number'])
# Расчёт времени прохождения модуля
module_df['module_time'] = (module_df['finish_date'] - module_df['start_date'])
# Исключение модулей нулевым и отрицательным временем прохождения (ошибки данных)
module_df = module_df[module_df['module_time'] > pd.to_timedelta('0 days +00:00:00.000000')]
# Расчёт минимального, максимального, среднего и медианного времён прохождения модуля
module_df['minimal_time'] = module_df['module_time']
module_df['maximal_time'] = module_df['module_time']
module_df['mean_time'] = module_df['module_time']
module_df['median_time'] = module_df['module_time']
# Переиндексирование
module_df.reset_index(inplace=True)
# Добавление расчитанных данных к рабочему датасету через цикл по курсам и цикл по модулям
module_df_ = pd.DataFrame([ ])
for title in module_df['course_title'].unique():
    title_mod_df = module_df[module_df['course_title'] == title]
    for number in title_mod_df['module_number'].unique():
        m_df_slice = title_mod_df[title_mod_df['module_number'] == number]
        m_df_slice['minimal_time'] = m_df_slice['module_time'].min()
        m_df_slice['maximal_time'] = m_df_slice['module_time'].max()
        m_df_slice['mean_time'] = m_df_slice['module_time'].mean()
        m_df_slice['median_time'] = m_df_slice['module_time'].median()
        module_df_ = module_df_.append(m_df_slice)
# Удаление лишних столбцов, затем получившихся дублей и переиндексация
module_df = module_df_.drop(columns=[
    'progress_id', 'start_date', 'finish_date', 'module_time']).drop_duplicates().reset_index(drop=True)
# Сортировка финального сета.
module_df.sort_values(by=['course_title','module_number'], inplace=True, ignore_index=True)
module_df
```

```python
# 1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение 
# количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать 
# по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то 
# считать по последнему дз в модуле).

# создание рабочего датасета путем среза основного сета по 'course_title', 'module_number', 'module_title', 
#'start_date', 'finish_date', 'progress_id' при условии наличия домашнего задания
cnvrsn_df = df[df['is_homework'] == True].loc[
    :,['course_title', 'module_number', 'module_title', 'start_date', 'finish_date', 'progress_id', 'status']]
# Удаление NA по столбцам 'course_title', 'module_number', 'start_date', 'finish_date'
cnvrsn_df.dropna(subset=['course_title', 'module_number', 'start_date'], inplace=True)
# Расчет времени начала модуля для каждого прогресса
cnvrsn_start = cnvrsn_df.groupby(by=[
    'progress_id', 'course_title', 'module_number', 'module_title'], as_index=False)['start_date'].min()
# Расчет количества прогрессов, начавших модуль
cnvrsn_start = cnvrsn_start.groupby(by=[
    'course_title', 'module_number', 'module_title'], as_index=False)['progress_id'].count()
# Переименование столбца 'progress_id' в 'start_quantity'
cnvrsn_start.rename(columns={'progress_id' : 'start_quantity'}, inplace=True)
# Расчет времени окончания модуля для каждого прогресса [cnvrsn_df.status == 'done']
cnvrsn_finish = cnvrsn_df.groupby(by=[
    'progress_id', 'course_title', 'module_number', 'module_title'], as_index=False)['finish_date'].max()
# Расчет количества прогрессов, окончивших модуль
cnvrsn_finish = cnvrsn_finish.dropna().groupby(by=[
    'course_title', 'module_number', 'module_title'], as_index=False)['progress_id'].count()
# Переименование столбца 'progress_id' в 'finish_quantity'
cnvrsn_finish.rename(columns={'progress_id' : 'finish_quantity'}, inplace=True)
#Слияние датафреймов с количеством начавших и количеством закончивших
cnvrsn_df = cnvrsn_start.merge(cnvrsn_finish, how='outer', on=['course_title', 'module_number', 'module_title'])
# Расчет конверсии студентов из одного модуля в другой
cnvrsn_df['conversion'] = 0
for course_title in cnvrsn_df['course_title'].unique():
    for i in cnvrsn_df[cnvrsn_df['course_title'] == course_title].index:
        if (i-1) in cnvrsn_df[cnvrsn_df['course_title'] == course_title].index:
            cnvrsn_df['conversion'][i] = cnvrsn_df['start_quantity'][i]/cnvrsn_df['finish_quantity'][(i-1)]*100
        else:
            cnvrsn_df['conversion'][i] = None
#Удаление NAN            
cnvrsn_df.dropna(inplace=True)
cnvrsn_df['finish_quantity'] = cnvrsn_df['finish_quantity'].astype('int32')
# Переиндексация
cnvrsn_df.reset_index(inplace=True, drop=True)
cnvrsn_df

#cnvrsn_start
```

```python
# 3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию 
# перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks 
# должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой 
# относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия 
#медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐

# Графики оформляю в 1 столбец
fig, axes = plt.subplots(
    int(course_for_plot['course_title'].count()),
    figsize=(8,120), 
    subplot_kw=dict(frame_on=False)
)
# Вывод заголовка для всего листа
fig.suptitle(
    'Конверсия перехода студентов из одного модуля в другой по курсам',
    x=-0.15,
    y=0.885, 
    fontsize='xx-large', 
    fontweight='bold'
)
# Генерация графиков
for row in range(int(course_for_plot['course_title'].count())):
    # Формирование датасета для конкретного курса
    increase_data = cnvrsn_df[cnvrsn_df.course_title == course_for_plot.course_title[row]]
    #Генерация меток модулей по шаблону "номер - название"
    yticks = increase_data.module_number.astype('str') + ' - ' + increase_data.module_title
    #Формирование меток оси X (если максимальное значение более 110%, то шкала и метки расчитываются исходя 
    #из максимального значения, иначе - шкала от 0 до 100 с шагом 10)
    if max(increase_data['conversion']) > 110:
        xticks = range(
            0, 
            math.ceil(max(increase_data['conversion'] / 10 + 1)) * 10, 
            math.ceil(max(increase_data['conversion'] / 100)) * 10
        )
    else:
        xticks = range(
            0, 
            110, 
            10
        )
    #Отрисовка линии медианного значения
    median_label_value = cnvrsn_median['median_conversion'][course_for_plot.course_title[row]]
    median_label = str('Медиана({:.2f} %)'.format(median_label_value))
    
    axes[row].barh(yticks, increase_data['conversion'], alpha=0.1) #Построение графика
    axes[row].axvline(x=median_label_value, ls='--', color='g', label=median_label) #Отрисовка линии медианного значения
    axes[row].set_xlabel('Конверсия, %') #Название оси X
    axes[row].set_ylabel('Номер и название модуля') #Название оси Y
    axes[row].set_title('Курс "{}"'.format(course_for_plot.course_title[row])) #Титул графика
    
    #Задание сетки по оси  Y и надписей отметок
    axes[row].set_yticks(yticks)
    axes[row].set_yticklabels(yticks)
    #изменение направления оси Y
    axes[row].invert_yaxis() 
    #Задание сетки по оси X и надписей отметок
    axes[row].set_xticks(xticks) 
    axes[row].set_xticklabels(xticks)
    
    axes[row].legend(loc='lower right') #Вывод легенды (значение медианы)

plt.show()
```
